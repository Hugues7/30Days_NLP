```python
# prompt: matcher le code postale du dataframe new_df avec le code_postal du dataframe  new_df2 si new_df[code_postale] == new_df[code_postal] alors on ajoute dans le daframe new_df le nom_departement  correspondant contenue le dataframe new_df2. NB: respecter l'ordre de colonne code postale du dataframe new_df

import pandas as pd

# Assuming df1 and df2 are already loaded as in the previous code
# df1=pd.read_csv('/content/datatourism.csv')
# df2=pd.read_csv('/content/20230823-communes-departement-region.csv')

if 'Code_postal_et_commune' in df1.columns:
    df1[['code_postale', 'commune']] = df1['Code_postal_et_commune'].str.split('#', n=1, expand=True)
    #new_df = df1[['code_postale', 'commune']].copy()
else:
    print("La colonne 'Code_postal_et_commune' n'existe pas dans le DataFrame.")

new_df2 = df2[['code_departement', 'nom_departement', 'code_postal']].copy()

# Convert 'code_postal' columns to string type to avoid comparison issues
df1['code_postale'] = df1['code_postale'].astype(str)
new_df2['code_postal'] = new_df2['code_postal'].astype(str)


# Merge the two dataframes based on 'code_postale'
df1 = pd.merge(df1, new_df2, left_on='code_postale', right_on='code_postal', how='left')

# Select the desired columns and reorder them
#final_df = newdf[['code_postale', 'commune', 'nom_departement']]

# Display the resulting DataFrame
df1
```
